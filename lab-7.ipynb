```python
import pandas as pd
import numpy as np
import random
```

**<h3>1) to create a dataframe from a dictionary and display it</h3>**

```python
data_frame2 = {"first Number": np.random.randint(1,50,4), "Second Number": np.random.randint(1,50,4)}

data_frame2 = pd.DataFrame(data_frame2, index=["a", "b", "c", "d"])
print("Data Frame :\n", data_frame2,sep="")
```

**<h3>2) to sort the DataFrame first by 'name' in ascending order</h3>**

```python
data_frame3 = {"Name":["yash","mihir","chintan","vansh","shubham","nayan"],"Enrollment":[22, 33, 29, 37, 30, 36]}
data_frame3 = pd.DataFrame(data_frame3)
print("Data Frame : \n",data_frame3,sep="")
data_frame3 = data_frame3.sort_values(by="Name",ascending=True)
print("\nData Frame : \n",data_frame3,sep="")
```

**<h3>3) to delete the one specific column from the DataFrame</h3>**

```python
data_frame4 = pd.DataFrame({"Name":["yash","chintan","vansh","shubham"],"Enrollment":[22, 29, 37, 30]})
print("Data Frame :\n",data_frame4,sep="")
data_frame4 = data_frame4.drop(["Enrollment"],axis=1)
print("Data Frame with Deleted column :\n",data_frame4,sep="")
```

**<h3>4) to write a DataFrame to CSV file using tab separator</h3>**

```python
data_frame5 = pd.DataFrame({"Student": ["yash", "mhir", "vansh", "nayan"],
"Enrollment": [22, 33, 37, 36]},columns =["Student", "Enrollment"])

print("Data Frame :\n",data_frame5,sep="")
 
data_frame5.to_csv("Students.csv", sep ="\t",index=False)
 
data_frame5 = pd.read_csv("Students.csv",sep="\t")
 
print("\nData from Students.csv :\n",data_frame5,sep="")
```
