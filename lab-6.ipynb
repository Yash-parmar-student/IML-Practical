```python
import numpy as np
import pandas as pd
import random
```

**<h3>1) to convert a NumPy array to a Pandas series.</h3>**

```python
pandas_series = pd.Series(np.random.randint(1,20,10))
print("Pandas Series :")
print(pandas_series)
```

**<h3>2) to convert the first column of a DataFrame as a Series</h3>**

```python
data_frame = pd.DataFrame({'Enrollment': [22, 29, 33], 'SPI': [8.43,  8.94, 8.86]})
print("Data Frame : \n",data_frame,sep="");
series = pd.Series(data_frame.iloc[:,0])
print("\nSeries : \n",series,sep="")
```

**<h3>3) to create the mean and standard deviation of the data of a given Series</h3>**

```python
series2 = pd.Series(np.random.randint(1,20,5))
print("Pandas Series :")
print(series2)
print("Mean : ",series2.mean())
print("Standard Deviation : ",series2.std())
```

**<h3>4) to sort a given Series</h3>**

```python
series2 = pd.Series(np.random.randint(1,20,5))
print("Pandas Series :")
print(series2)
print("\nsorted Series : \n",series2.sort_values(ascending=True),sep="")
```
