**<h3>1) to add, subtract, multiply, divide arguments element-wise</h3>**

```python
array1 = np.arange(5)
plus_array = array1 + 4
print("array 1 : ",array1,end="\n\n")
print("result array with elements plus by 4 : ",plus_array,end="\n\n")

divide_array = plus_array // 2
print("result array with elements divide by 2 : ",divide_array,end="\n\n") 

minus_array = divide_array - 1
print("result array with elements minus by 1 : ",minus_array,end="\n\n") 

multiply_array = minus_array * 5
print("result array with elements mulitply by 1 : ",multiply_array,end="\n\n") 
```

**<h3>2) to round elements of the array to the nearest integer</h3>**

```python
from random import random

array2 = np.random.uniform(1.1,20.9,5)
print("Main array : ",array2)
array2 = np.round(array2)
print("Array with nearest integer : ",array2.astype(int))
```

**<h3>3) to calculate mean across dimension, in a 2D numpy array</h3>**

```python
array3 = np.random.uniform(1,20,4).reshape((2, 2))
print("Main array : ",array3)
print("Mean of array : ",array3.mean(),end="\n\n")

row_mean = np.mean(array3,axis=1)
print("row 1 mean : ",row_mean[0])
print("row 2 mean : ",row_mean[1],end="\n\n")

col_mean = np.mean(array3,axis=0)
print("column 1 mean : ",col_mean[0])
print("column mean : ",col_mean[1])
```

**<h3>4) to calculate the difference between neighboring elements, element-wise of a given array</h3>**

```python
array4 = np.random.uniform(1,20,5)
print("Main array : ",array4)

array4 = np.diff(array4)
print("diffrence between neighboring element of array : ",array4)
```
