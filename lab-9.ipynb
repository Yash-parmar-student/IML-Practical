**<h3>1) Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates</h3>**

```python
'''
Name: Parmar Yash Vindobhai
Enrollment: 216270307022
'''

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../input/nifty50-stock-market-data/AXISBANK.csv')

df.set_index('Date', inplace=True)

start_date = '2000-01-01'
end_date = '2000-01-20'

df = df.loc[start_date:end_date]

up = df[df.Close >= df.Open]
  
down = df[df.Close < df.Open]
  
width = .5
width2 = .1
  
plt.bar(up.index, up.Close-up.Open, width, bottom=up.Open,color="red")
plt.bar(up.index, up.High-up.Close, width2, bottom=up.Close,color="red")
plt.bar(up.index, up.Low-up.Open, width2, bottom=up.Open,color="blue")
  
plt.bar(down.index, down.Close-down.Open, width, bottom=down.Open,color="green")
plt.bar(down.index, down.High-down.Open, width2, bottom=down.Open,color="blue")
plt.bar(down.index, down.Low-down.Close, width2, bottom=down.Close,color="green")

plt.xticks(rotation='vertical')
plt.title('Stock Prices of Axis Bank')
plt.xlabel('Date')
plt.ylabel('Price/Volume')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(12, 3))
plt.plot(df.index, df.Volume, label='Volume', color='blue')

plt.xticks(rotation='vertical')
plt.title('Volume')
plt.xlabel('Date')
plt.ylabel('Price/Volume')
plt.grid(True)
plt.legend()
plt.show()
```
